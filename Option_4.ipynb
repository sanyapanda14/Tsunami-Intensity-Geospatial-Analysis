# Methodology

To examine the relation between the occurences of tsunamis, earthquakes, and volcanoes, we use Pandas to load the dataset of the events into a dataframe, read information from the file 'data/tsevent.csv', extract the required data of tsunami events occured after 2000 and not by a questionable earthquake from the loaded dataframe and plot them on the geospatial plot generated by using Cartopy and Matplotlib. Furthermore, we extend our analysis by importing and plotting tsunamis caused by earthquakes and volcanoes separately. This is followed by generating histograms presenting the intensity of tsunamis caused by volcanoes, earthquakes, and both (volcanoes and earthquakes). 


## 1. Loading the dataset of Tsunami events into a Pandas dataframe

````python
## 1. Code to load dataset into the dataframe

# To begin with, we import the Pandas dataframe and represent it using pd for ease of usage in writing further code
import pandas as pd 

# We read the data from file 'data/tsevent.csv' using the function 'read_csv' within the Pandas package
if 'google.colab' in str(get_ipython()):
    data = pd.read_csv('https://raw.githubusercontent.com/loftytopping/DEES_programming_course/master/data/tsevent.csv')
else:
    data = pd.read_csv("data/tsevent.csv")
    
#The '.' operator is used to access the attributes belonging to different object classes.
# We preview the first 5 lines of the loaded data using '.head()' 
data.head(5)

#### ------------------------------------
````

## 2. Extracting data from the loaded dataframe and creating a plot of locations of events.

````python
# 2. Code to select required data from the dataframe and plot it on the map.

# We import Cartopy and Matlpotlib as ccrs and plt respectively.
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# We create a canvas comprising of the map on which we shall plot our locations.
fig = plt.figure(figsize=(15,15))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

# We extract the 'LATITUDE' and 'LONGITUDE' columns and then their .values as follows.
Latitude = data['LATITUDE'].values 
Longitude = data['LONGITUDE'].values 

# We plot the extracted data points on the map.  
plt.scatter(Longitude, Latitude,
         color='blue', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
#
````

###                   Figure 1: Map with geospatial plots of Tsunami occurences 

## 3. Plotting tsunami events occured after 2000 and not by a questionable earthquake


````python
#Plotting tsunamis 
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,15))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

Latitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LATITUDE'].values  # Extract the 'Latitude' column and then its .values
Longitude = data[(data['YEAR']>2000) & (data['CAUSE_CODE']!=2)]['LONGITUDE'].values   # We can do the same with Longitude values
# <<------------------------------------------------------------->>

plt.scatter(Longitude, Latitude,
         color='black', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
#
````

### Figure 2: Geospatial plot of tsunamis occured after 2000 by a non-questionable earthquake

## 4. Analysing the plots of tsunamis caused by volcanoes and earthquakes 

````python
# 4. Plotting and analysing the relation between volcano and earthquake induced tsunamis. 
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, axs = plt.subplots(nrows=1,ncols=2,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(15,15))


axs[0].stock_img()
axs[1].stock_img()

Latitude = data[data['CAUSE_CODE']==6]['LATITUDE'].values # Extract the 'Latitude' column and then its .values
Longitude = data[data['CAUSE_CODE']==6]['LONGITUDE'].values # We can do the same with Longitude values
Latitude1 = data[data['CAUSE_CODE']==1]['LATITUDE'].values # Extract the 'Latitude' column and then its .values
Longitude1 = data[data['CAUSE_CODE']==1]['LONGITUDE'].values # We can do the same with Longitude values

axs[0].scatter(Longitude, Latitude,
         color='red', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )
axs[1].scatter(Longitude1, Latitude1,
            color='green', linewidth=2, s=9, marker='o',
            transform=ccrs.PlateCarree()
           )

plt.show()



#---------------------------------------------
````

### Figure 3: Geospatial plot of tsunamis occured due to volcanic eruptions

## 5. Plotting tsunamis caused by earthquakes

````python
#Plotting the tsunamis due to earthquakes
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,15))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

Latitude = data[data['CAUSE_CODE']==1]['LATITUDE'].values # Extract the 'Latitude' column and then its .values
Longitude = data[data['CAUSE_CODE']==1]['LONGITUDE'].values # We can do the same with Longitude values

plt.scatter(Longitude, Latitude,
         color='brown', linewidth=2, s=9, marker='o',
         transform=ccrs.PlateCarree(),
         )

plt.show()
#---------------
````
